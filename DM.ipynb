```python
# import the pandas library for data manipulation
# import the MinMaxScaler class from sklearn.preprocessing for scaling numerical data

import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

load the data from Excel File (FoodMart Dataset) , display the first 5 Rows in it

```python
df = pd.read_excel('StoresData.xlsx')
df.head()
```

drop specified columns from the pandas DataFrame df since they are alternative columns

display the first 5 Rows in it after update

```python
df = df.drop(columns=["Sundays", "Mng-Sex", "HomeDel", "Location", "State"])
df.head()
```

```python
# Columns Definition :

# Store No. :  Store ID
# Sales $m : Sales in million dollars
# Wages $m : Wages in million dollars
# No. Staff : Number of staff
# Loc'n (Num) : Location (numeric encoded)
# State (Num) : State (numeric encoded)
# Age (Yrs) : Store age (in years)
# GrossProfit : Profit after paying for the goods
# Adv.$'000 : Advertising budget in thousands
# Competitors : Number of competitors
# Sundays (Num) : Open on Sundays (0/1)
# Mng-Sex (Num) : Manager gender (0/1)
# Mng-Age : Manager age
# Mng-Exp : Manager experience (years)
# Mng-Train : Manager training score or level
# Union% : Percentage of staff unionized
# Car Spaces : Number of parking spaces
# HomeDel (Num) : Home delivery available (0/1)
# Basket:2013 : Basket size or sales in 2013
# Basket:2014 : Basket size or sales in 2014
```

```python
# A MinMaxScaler was initialized and used to scale
# the 'Sales $m' and 'Adv.$'000' columns of DataFrame df to the range [0, 1]

scaler = MinMaxScaler()
df[['Sales $m', "Adv.$'000"]] = scaler.fit_transform(df[['Sales $m', "Adv.$'000"]])
print(df.head)
```

```python
# Outliers were identified and removed from the specified numerical columns of DataFrame df
# using the IQR method, resulting in df_cleaned, and the shapes of both DataFrames were printed

numerical_columns = ['Sales $m', 'Wages $m', 'Adv.$\'000', 'No. Staff', 'GrossProfit']
Q1 = df[numerical_columns].quantile(0.25)
Q3 = df[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_mask = (df[numerical_columns] < lower_bound) | (df[numerical_columns] > upper_bound)
df_cleaned = df[~outliers_mask.any(axis=1)]
print("Original df:", df.shape)
print("Cleaned df:", df_cleaned.shape)
```

```python
# Split the data into training and testing sets

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# predicts the probability of each sample in the testing data
# belonging to each class of the simple logistic regression model

y_pred_proba1 = model_simple.predict_proba(x_test[['Car Spaces']])
y_pred_proba1
```

```python
# K-Medoids Clustering
```

```python
# The 'Sales $m' and 'No. Staff' columns were selected from df_cleaned
# and their values were printed as a NumPy array

from sklearn_extra.cluster import KMedoids

column= ["Sales $m", "No. Staff"]
two_numeric_column=df_cleaned[column].values
print(two_numeric_column)
```

```python
# K-Medoids model with 3 clusters was fitted to the 'Sales $m' and 'No. Staff' columns
# the resulting labels, medoid indices, cluster centers, and data points within each cluster were printed

k = 3
kmedoids = KMedoids(n_clusters=k, random_state=42)
kmedoids.fit(two_numeric_column)

labels = kmedoids.labels_
medoids = kmedoids.medoid_indices_
clusters=kmedoids.cluster_centers_

print("Labels :", labels, "\n")
print("Medoids :", medoids, "\n")
print("Clusters :", clusters, "\n")

for j in range (k):
    for i in range (len(two_numeric_column)):
        if (kmedoids.labels_[i]==j):
            x=two_numeric_column[i]
            print("cluster", j , ":", x)
```

```python
# Doing scatter plot represents the K-Medoids clustering of stores based on 'Sales $m' and 'No. Staff'
# with points colored by cluster and medoids marked with red 'X's

plt.figure(figsize=(10, 6))
plt.scatter(two_numeric_column[:, 0], two_numeric_column[:, 1],
           c=labels, cmap='viridis', alpha=0.7, label='Stores')
plt.scatter(two_numeric_column[medoids, 0], two_numeric_column[medoids, 1],
           c='red', marker='X', s=200, label='Medoids')
plt.xlabel('Sales ($m)')
plt.ylabel('Number of Staff')
plt.title('K-Medoids Clustering of Stores')
plt.legend()
plt.grid(True)
plt.show()
```

Hierarchical Clustering Dendrogram

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler

column= ["Sales $m","Wages $m","No. Staff"]
hir_data=df_cleaned[column]

scaler = StandardScaler()
scaled_data = scaler.fit_transform(hir_data)

plt.figure(figsize=(6, 6))

plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c='r')

for i in range(scaled_data.shape[0]):
    plt.annotate(str(i), xy=(scaled_data[i, 0], scaled_data[i, 1]), xytext=(3, 3), textcoords='offset points')

plt.xlabel('x coordinate')
plt.ylabel('y coordinate')
plt.title('Scatter Plot of the data')
plt.grid()
plt.show()

```

```python
linked = linkage(scaled_data, method='ward')



# single_linkage=linkage(linked,method='single',metric='euclidean')

plt.figure(figsize=(12, 8))
plt.title("Hierarchical Clustering Dendrogram")
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```
